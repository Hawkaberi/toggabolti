## Combinations

```python
# Initialize a dictionary to store DataFrames for each season
season_combination_dfs = {}

# Group dagadict by season
for season in set(value['season'] for value in dagadict.values()):
    # Filter dagadict to only include entries for the current season
    season_data = {date: values for date, values in dagadict.items() if values['season'] == season}

    # Extract all unique names across lid1 and lid2 for the current season
    all_lid1_names = [name for sublist in [v['lid1'] for v in season_data.values()] for name in sublist]
    all_lid2_names = [name for sublist in [v['lid2'] for v in season_data.values()] for name in sublist]
    
    # Get all combinations of two names
    all_name_combinations = list(combinations(set(all_lid1_names + all_lid2_names), 2))
    
    # Initialize dictionary to store counts for each combination in this season
    name_combination_counts = {combination: {'total': 0, 'with_sigurvegari': 0} for combination in all_name_combinations}
    
    # Count occurrences of each combination in 'lid1' and 'lid2', including those with 'sigurvegari'
    for date, values in season_data.items():
        lid1_names = set(values['lid1'])
        lid2_names = set(values['lid2'])
        sigurvegari = values['sigurvegari']
        
        # Count occurrences of each combination in 'lid1' and 'lid2'
        for combination in all_name_combinations:
            if combination[0] in lid1_names and combination[1] in lid1_names:
                name_combination_counts[combination]['total'] += 1
                if sigurvegari == 'lid1':
                    name_combination_counts[combination]['with_sigurvegari'] += 1
            
            if combination[0] in lid2_names and combination[1] in lid2_names:
                name_combination_counts[combination]['total'] += 1
                if sigurvegari == 'lid2':
                    name_combination_counts[combination]['with_sigurvegari'] += 1
    
    # Create a DataFrame for this season
    df_season = pd.DataFrame(name_combination_counts).T.reset_index()
    df_season.columns = ['leikmadur1', 'leikmadur2', 'heild', 'sigrar']
    df_season['sigurhlutfall'] = df_season['sigrar'] / df_season['heild']
    
    # Sort and reset index
    df_season = df_season.sort_values(by='heild', ascending=False).reset_index(drop=True)
    
    # Store the DataFrame in the dictionary with the season number as the key
    season_combination_dfs[season] = df_season

# Now each season's combinations are in the `season_combination_dfs` dictionary
# You can access each season's DataFrame by season number
# For example, to view season 1:
season_combination_dfs[2]
```

## Glicko - 2 og Elo

```python
import math

# Elo constants
initial_elo_rating = 1200

# Function to calculate the E value in Elo
def E_elo(r, ri):
    return 1 / (1 + 10 ** ((ri - r) / 400))

def get_k_factor(player):
    games = player['games']

    slope = (10 - 80) / (100 - 0)  # = -70/90 = -7/9 ≈ -0.778
    return 80 + slope * (games - 0)

def calculate_rating_change(player, results):
    K = get_k_factor(player)
    total_delta = 0
    num_opponents = len(results)
    multiplier = min(1 + 0.1 * (player['win_streak'] - 1), 2)

    for match in results:
        opponent_rating = match['opponent']['elo_rating']
        score = match['score']  # 1 for win, 0 for loss
        expected_score = E_elo(player['elo_rating'], opponent_rating)
        base_delta = K * (score - expected_score)
        
        adjusted_delta = base_delta * multiplier
        total_delta += adjusted_delta
    
    # Normalize so total rating change doesn't inflate with number of opponents
    normalized_change = (total_delta*3) / num_opponents if num_opponents > 0 else 0

    avg_opponent_rating = sum(r['opponent']['elo_rating'] for r in results) / num_opponents if num_opponents > 0 else player['elo_rating']
    avg_score = sum(r['score'] for r in results) / num_opponents if num_opponents > 0 else 0

    return {
        'leikmadur': player['leikmadur'],
        'elo_change': normalized_change,
        'avg_opponent_rating': avg_opponent_rating,
        'k_factor': K,
        'avg_score': avg_score,
        'win_streak': player['win_streak'],
        'loss_streak': player['loss_streak']
    }

def update_all_players(results, inactive_players, date=None):
    updated_players = []
    rating_changes = []

    # First pass: calculate rating changes based on original ratings
    for result in results:
        player = result['player']
        match_results = result['results']
        change_data = calculate_rating_change(player, match_results)
        rating_changes.append((player, change_data))
        player['isactive'] = True

    # Second pass: apply rating changes
    for player, change_data in rating_changes:
        if player['leikmadur'] == 'Álfgrímur':
            print(f"\n=== Match for Þórður on {date} ===")
            print(f"Games before: {player['games']}")
            print(f"ELO before: {player['elo_rating']:.2f}")
            print(f"Avg opponent ELO: {change_data['avg_opponent_rating']:.2f}")
            print(f"Avg score: {change_data['avg_score']:.2f}")
            print(f"K-factor used: {change_data['k_factor']:.2f}")
            print(f"ELO change: {change_data['elo_change']:.2f}")
            print(f"Win streak: {player['win_streak']}")
            print(f"Loss streak: {player['loss_streak']}")
        
        player['elo_rating'] += change_data['elo_change']
        player['games'] += 1
        updated_players.append(player)

    # Preserve inactive players
    updated_players.extend(inactive_players)

    return updated_players
```

```python
leikmannalist = []
for leikmadur in df.columns[1:]:
    player = {
        'leikmadur': leikmadur,
        'elo_rating': initial_elo_rating,
        'games': 0,
        'sigrar': 0,
        'jafntefli': 0,
        'top': 0,
        'win_streak': 0,
        'loss_streak': 0,
        'isactive': False
    }
    leikmannalist.append(player)

leikmanna_dict = {player['leikmadur']: player for player in leikmannalist}

for date, values in dagadict.items():
    results = []

    if values['sigurvegari'] == 'lid1':
        winners = values['lid1']
        losers = values['lid2']

        for player_name in winners:
            player = leikmanna_dict[player_name]
            player['sigrar'] += 1
            
            # Update streaks
            player['win_streak'] += 1
            player['loss_streak'] = 0

            results.append({
                'player': player,
                'results': [{'opponent': leikmanna_dict[opp], 'score': 1} for opp in losers]
            })

        for player_name in losers:
            player = leikmanna_dict[player_name]
            player['top'] += 1

            # Update streaks
            player['loss_streak'] += 1
            player['win_streak'] = 0

            results.append({
                'player': player,
                'results': [{'opponent': leikmanna_dict[opp], 'score': 0} for opp in winners]
            })

    elif values['sigurvegari'] == 'lid2':
        winners = values['lid2']
        losers = values['lid1']

        for player_name in winners:
            player = leikmanna_dict[player_name]
            player['sigrar'] += 1

            # Update streaks
            player['win_streak'] += 1
            player['loss_streak'] = 0

            results.append({
                'player': player,
                'results': [{'opponent': leikmanna_dict[opp], 'score': 1} for opp in losers]
            })

        for player_name in losers:
            player = leikmanna_dict[player_name]
            player['top'] += 1

            # Update streaks
            player['loss_streak'] += 1
            player['win_streak'] = 0

            results.append({
                'player': player,
                'results': [{'opponent': leikmanna_dict[opp], 'score': 0} for opp in winners]
            })

    elif values['sigurvegari'] == 'tie':
        team1 = values['lid1']
        team2 = values['lid2']

        for player_name in team1:
            player = leikmanna_dict[player_name]
            player['jafntefli'] += 1

            # Reset streaks on tie
            player['win_streak'] = 0
            player['loss_streak'] = 0

            results.append({
                'player': player,
                'results': [{'opponent': leikmanna_dict[opp], 'score': 0.5} for opp in team2]
            })

        for player_name in team2:
            player = leikmanna_dict[player_name]
            player['jafntefli'] += 1

            # Reset streaks on tie
            player['win_streak'] = 0
            player['loss_streak'] = 0

            results.append({
                'player': player,
                'results': [{'opponent': leikmanna_dict[opp], 'score': 0.5} for opp in team1]
            })

    # Update all players
    # Combine both teams' ids into a single list
    active_player_ids = values['lid1'] + values['lid2']

    # Get the list of player dictionaries who are not on either team
    inactive_players = [player for player in leikmannalist if player['leikmadur'] not in active_player_ids]

    leikmannalist = update_all_players(results, inactive_players, date=date)

    # Convert the list of dictionaries to a DataFrame
    leikmanna_df = pd.DataFrame(leikmannalist)

    # Convert the DataFrame back to a list of dictionaries
    leikmannalist = leikmanna_df.to_dict(orient='records')

    leikmanna_dict = {player['leikmadur']: player for player in leikmannalist}

    players_df = pd.DataFrame(leikmannalist)

    dagadict[date]['leikmenn'] = players_df

players_df['win_percentage'] = (players_df['sigrar'] / players_df['games']) * 100
players_df['tie_percentage'] = (players_df['jafntefli'] / players_df['games']) * 100
players_df['lose_percentage'] = (players_df['top'] / players_df['games']) * 100

players_dict = players_df.reset_index().to_dict(orient='records')

# Save player stats to a JSON file
with open('./player_stats.json', 'w') as f:
    json.dump(players_dict, f)

display(players_df.sort_values('elo_rating', ascending=False))
display(players_df[players_df['games']>4].sort_values('elo_rating', ascending=False))
```
